```python
import os, json
import matplotlib.pyplot as plt

path = '.\\sample_label\\'
```

```python
dw, dh = [], []

for json_name in os.listdir(path) :
    with open(path+json_name, 'r', encoding='utf8') as f :
        label = json.load(f)

        for box in label["bbox"] :
            dw.append(box["x"][-1] - box["x"][0])
            dh.append(box["y"][-1] - box["y"][0])
```

```python
f, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].set_title('Histogram of Box Height')
ax[0].hist(dh, range=(0, 300), bins=50)

ax[1].set_title('Histogram of Box Weight')
ax[1].hist(dw, range=(0, 600), bins=100)

plt.show()
```
