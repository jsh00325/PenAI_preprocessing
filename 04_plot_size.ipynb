가로세로 뽑기

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

path_1 = './female/train/'
path_2 = './male/train/'

dw =[]
dh =[]
for detail_file_name in os.listdir(path_1):
    im = Image.open(path_1+ detail_file_name)
    width, height = im.size
    dw.append(width)
    dh.append(height)

for detail_file_name in os.listdir(path_2):
    im = Image.open(path_2+ detail_file_name)
    width, height = im.size
    dw.append(width)
    dh.append(height)


```

```python
#히스토그램을 가르키는 ax 
f, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].set_title('Histogram of Box Height')
ax[0].hist(dh, range=(0, 300), bins=50)

ax[1].set_title('Histogram of Box Weight')
ax[1].hist(dw, range=(0, 600), bins=50)

plt.show()
```
